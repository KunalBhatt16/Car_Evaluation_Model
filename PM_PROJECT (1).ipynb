```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import sklearn as skl
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_excel('CARS_DATA.xlsx')
```

```python
data = pd.DataFrame(data)
```

```python
data.head(10)
```

```python
# Checking for null values
print(data.isnull().sum())
```

```python
# Removing the rows with missing values and duplicate values
data.drop_duplicates(inplace = True)
data.dropna(inplace = True)
```

```python
# uniqueTransmission = data['transmission'].nunique()
# print(uniqueTransmission)
# uniquefuelType = data['fuelType'].nunique()
# print(uniquefuelType)

data.transmission.unique()
```

```python
data.fuelType.unique()
```

```python
data.loc[data['transmission'] == 'Manual','transmission'] = 1
data.loc[data['transmission'] == 'Automatic','transmission'] = 2
data.loc[data['transmission'] == 'Semi-Auto','transmission'] = 3
data.loc[data['transmission'] == 'Other','transmission'] = 4
```

```python
data.loc[data['fuelType'] == 'Petrol','fuelType'] = 1
data.loc[data['fuelType'] == 'Hybrid','fuelType'] = 2
data.loc[data['fuelType'] == 'Diesel','fuelType'] = 3
data.loc[data['fuelType'] == 'Electric','fuelType'] = 4
data.loc[data['fuelType'] == 'Other','fuelType'] = 5
```

```python
data
```

```python
data.describe()
```

```python
data['transmission'] = data['transmission'].astype(int)
data['fuelType'] = data['fuelType'].astype(int)
```

```python
Label = data['price']
Features = data[['year','mileage','tax','mpg','engineSize','transmission','fuelType']]

x_train,x_test,y_train,y_test = train_test_split(Features,Label,test_size = 0.3, random_state = 10)

print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# DEcision Tree
model = DecisionTreeRegressor(max_depth = 10)
model.fit(x_train,y_train)
```

```python
prediction_decision = model.predict(x_test)
accuracy_decision = r2_score(y_test,prediction_decision)
print(accuracy_decision)
```

```python
result_tree = pd.DataFrame({
    'Actual Price': y_test.values,
    'Predicted Price': prediction_decision,
    'residual':y_test.values - prediction_decision
})

result_tree.head(10)
```

```python
import graphviz
from sklearn import tree
```

```python
DTviz = tree.export_graphviz(model, out_file=None,
                                feature_names=Features.columns,
                                class_names='class',
                                filled=True,
                                proportion=True)  # Adjusts the size of the nodes

# Draw graph
graph = graphviz.Source(DTviz, format="png")

# graph.render("tree", format="png", cleanup=True)
graph
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# K Nearest Neighbors
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(x_train, y_train)
```

```python
prediction_knn = knn.predict(x_test)
accuracy_knn = r2_score(y_test,prediction_knn)
print(accuracy_knn)
```

```python
result_knn = pd.DataFrame({
    'Actual Price': y_test.values,
    'Predicted Price': prediction_knn
})

result_knn.head(10)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
# GradientBoosting
nb_model = GradientBoostingRegressor()
nb_model.fit(x_train,y_train)
```

```python
prediction_gnb = nb_model.predict(x_test)
accuracy_gnb = r2_score(y_test,prediction_gnb)
print(accuracy_gnb)
```

```python
result_gnb = pd.DataFrame({
    'Actual Price': y_test.values,
    'Predicted Price': prediction_gnb
    })

result_gnb.head(10)
```
